# Regresion Lineal Simple

```python
# Se van a emplear Datos generados:

X= np.array([5,15,25,35,45,55]).reshape((-1,1))

Y= np.array([5,20,14,32,22,38])

print (f"Datos X:\n{X}")

print (f"Datos Y:\n{Y}")
```

```python
# Scatter Plot

plt.figure(figsize=[5,5])

plt.scatter(x=X,y=Y)

plt.xlabel("Datos X")
plt.ylabel("Datos Y")

plt.show()
```

```python
#  Algoritmo Regresion Lineal Simple

modelo_regresion_lineal= LinearRegression().fit(X, Y)

r_sq= modelo_regresion_lineal.score( X, Y)

intercepcion_recta= modelo_regresion_lineal.intercept_

pendiente_recta= modelo_regresion_lineal.coef_

print(f"Coeficiente de Correlacion:{r_sq}")

print (f"Ecuación de la Recta:\nY= {pendiente_recta}X + {intercepcion_recta}")

```

```python
# Se generan otra serie de datos X, para  ejecutar en el modelo Regresión Líneal

X_nuevo= np.array([6,8,15,25,30,48]).reshape((-1, 1))

Y_pred= modelo_regresion_lineal.predict(X_nuevo)

print (f"Predicciones de Y:\n{Y_pred}")
```

```python
Y_pred_otro= modelo_regresion_lineal.intercept_+modelo_regresion_lineal.coef_*X_nuevo # Otro método para el cálculo de Y empleando; la intercepdión y la pendiente

print (f"Predicciones de Y:\n{Y_pred_otro}")
```

```python
# Generar Recta 
predict_space= np.linspace(min(X),max(X)).reshape((-1,1))
predicted= modelo_regresion_lineal.predict(predict_space)

#Plot Regression Line and Scatter
plt.plot(predict_space, predicted, color="black", linewidth=2)

plt.scatter(x=X, y=Y)
plt.scatter(x=X_nuevo, y=Y_pred)

plt.xlabel("Datos X")

plt.ylabel("Datos Y")

plt.show
print (f"R_Cuadrada:\n{modelo_regresion_lineal .score(X,Y)}")
```

# Multiple Regresión Lineal

```python
# Se van a emplear Datos generados:

x_regml= [[0,1],[5,1],[15,2],[25,5],[35,11],[45,15],[55,34],[60,35]]

y_regml= [4,5,20,14,32,22,38,43]

X_regmL,Y_regmL= np.array(x_regml), np.array(y_regml)

print(f"Datos X:\n{X_regmL}")

print(f"Datos Y:\n{Y_regmL}")
```

```python
# Extracción de los datos de las tuplas generadas

tupla= []

dato1= 0
dato2=1


for n in range(0,8):
    
    ext_tupla= X_regmL[n]
        
    tupla.append(ext_tupla)
    
tupla0, tupla1, tupla2, tupla3, tupla4, tupla5, tupla6, tupla7= tupla[0], tupla[1], tupla[2], tupla[3], tupla[4], tupla[5], tupla[6], tupla[7]

x1= np.array([tupla0[dato1], tupla1[dato1], tupla2[dato1], tupla3[dato1], tupla4[dato1], tupla5[dato1], tupla6[dato1], tupla7[dato1]]).reshape((-1, 1))

x2= np.array([tupla0[dato2], tupla1[dato2], tupla2[dato2], tupla3[dato2], tupla4[dato2], tupla5[dato2], tupla6[dato2], tupla7[dato2]]).reshape((-1, 1))
      
   
#Plot Scatter

plt.figure(figsize=[5,5])

plt.scatter(x=x1,y=Y_regmL, color="red")
plt.scatter(x=x2, y=Y_regmL, color="blue")

plt.title("Grafica con los Puntos Generados")
plt.xlabel("Datos X")
plt.ylabel("Datos Y")

plt.show()
   
```

```python
X_regmL
```

```python
# Algoritmo Multiple Regresion Lineal.

modelo_regresionL_multiple= LinearRegression().fit(X_regmL,Y_regmL)

r_sq_RLM= modelo_regresionL_multiple.score(X_regmL, Y_regmL)

intercepcion_recta_RLM= modelo_regresionL_multiple.intercept_

pendiente_recta_RLM= modelo_regresionL_multiple.coef_

print(f"Coeficiente de Correlacion:{r_sq_RLM}")

print (f"Ecuación de la Recta:\nY= {pendiente_recta_RLM}X + {intercepcion_recta_RLM}")
```

```python
# Se generan otra serie de datos X, para  ejecutar en el modelo Regresión Líneal

X_nuevo_RLM= np.array([[1,3], [60,34], [23,40], [3,60], [55, 26], [22, 18], [3, 0], [0, 10]])

Y_pred_RML= modelo_regresionL_multiple.predict(X_nuevo_RLM)

print (f"Y Cálculada:\n{Y_pred_RML}")

# De igual manera que en el modelo de Regresión Lineal Simple se puede emplear la ecuación de la recta para calcular Y, la unica diferecia seria la siguiente:
# Y_pred_RML= intercepcion_recta_RML+np.sum(pendiente_recta_RML*X, axis=1), # Se emplea np.sum debido a que estamos tabajando con Matrices de (1-2) o tuplas de números.
```

```python
# Extracción de los datos de las tuplas generadas

tupla= []

dato1= 0
dato2=1


for n in range(0,8):
    
    ext_tupla= X_nuevo_RLM[n]
        
    tupla.append(ext_tupla)
    
tupla_0, tupla_1, tupla_2, tupla_3, tupla_4, tupla_5, tupla_6, tupla_7= tupla[0], tupla[1], tupla[2], tupla[3], tupla[4], tupla[5], tupla[6], tupla[7]

x1_n= np.array([tupla_0[dato1], tupla_1[dato1], tupla_2[dato1], tupla_3[dato1], tupla_4[dato1], tupla_5[dato1], tupla_6[dato1], tupla_7[dato1]])

x2_n= np.array([tupla_0[dato2], tupla_1[dato2], tupla_2[dato2], tupla_3[dato2], tupla_4[dato2], tupla_5[dato2], tupla_6[dato2], tupla_7[dato2]])
      
```

```python
# Para Graficar Regresiones Multiples se tienen que ocupar graficos 3D debido a las dimensiones de los datos ya que representan a un plano (x, y, z) no a una recta (x, y).

# Regresion Linean Multiple 

z= modelo_regresionL_multiple.intercept_+modelo_regresionL_multiple.coef_[0]*x1+ modelo_regresionL_multiple.coef_[1]*x2

# Z es la regresion Lineal Multiple
```

```python
fig=plt.figure(figsize=[5,5])

ax= fig.subplots()

ax.scatter(x=x1,y=Y_regmL, color="red")
ax.scatter(x=x2, y=Y_regmL, color="blue")
ax.scatter(x= x1_n, y=  Y_pred_RML, color= "Orange")
ax.scatter(x= x2_n, y=Y_pred_RML, color= "Yellow")

ax.plot(x1,z, color="black")
ax.plot(x2, z, color="black")

plt.title("Regresión Líneal Multiple")
plt.xlabel("Datos X")
plt.ylabel("Datos Y")

plt.show()
```

```python
fig= plt.figure(figsize=(10,10))

ax= fig.add_subplot(projection="3d")

ax.scatter(x1, x2, Y_regmL, color= "red")

ax.scatter(x1_n, x2_n, Y_pred_RML, color="green")

ax.plot(x1, x2, z)

plt.title("Regresion Lineal Multiple")
plt.xlabel("Datos x1")
plt.ylabel("Datos x2")


fig.show()
```

# Regresión Polinomial

```python
# Generar Datos 

X, Y = make_regression(n_samples= 50, n_features= 2, noise= 20, random_state= 42)

X_P= np.array(X)

Y_P= np.array(Y)
```

```python
#Plot Scatter

plt.figure(figsize=[5,5])


plt.scatter(x=X_P[:,1],y=Y_P, color="red")
plt.scatter(x= X_P[:, 0], y=Y_P, color="blue")

plt.title("Grafica con los Puntos Generados")
plt.xlabel("Datos X")
plt.ylabel("Datos Y")

plt.show()
```

```python
# Algoritmo Regresion Lineal Polinomial

transformer= PolynomialFeatures(degree=2, include_bias=False)

transformer.fit(X_P)

X_polinomial= transformer.transform(X_P)

modelo_RLPolinomial= LinearRegression().fit(X_polinomial, Y_P)

# Datos de la Regresión

r_sq_RLPolinomial= modelo_RLPolinomial.score(X_polinomial, Y_P)
Intercepcion_RLPolinomial= modelo_RLPolinomial.intercept_
pendiente_RLPolinomial= modelo_RLPolinomial.coef_

print (f"Coeficiente de Correlación:{r_sq_RLPolinomial}")
print (f"Ecuación de la Recta: Y= X{pendiente_recta_RLM}+ {Intercepcion_RLPolinomial}")
```

```python
#Datos Nuevos

X_NLRP, Y_NLRP = make_regression(n_samples= 50, n_features= 5, noise= 20, random_state= 30)
 
# Se deben generar datos con n_features=5

X_P_NRLP= np.array(X_NLRP)

# Predicción Datos

Y_PRED_LRP= modelo_RLPolinomial.predict(X_P_NRLP)

print (f"Predicción Y:\n{Y_PRED_LRP}")
```
